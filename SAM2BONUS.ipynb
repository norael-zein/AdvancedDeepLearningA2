```python
#Perform segmentation on all images
from ultralytics import SAM
import os
import torch
import glob

#Load SAM model
model = SAM("sam2.1_t.pt")

device = "cuda" if torch.cuda.is_available() else "cpu"

#Output directory
output_dir = "./segmented_results"
os.makedirs(output_dir, exist_ok=True)
```

```python
image_paths = glob.glob("./dataset/test/images/*")
for image_path in image_paths:
    image_name = os.path.basename(image_path)
    print(f"Segmenting: {image_name}")

    results = model.predict(source=image_path, device=device)

    output_path = os.path.join(output_dir, image_name)
    results[0].save(filename=output_path)
```

```python
from ultralytics import SAM
import cv2
import os
import torch

#Load model
model = SAM("sam2.1_b.pt")
device = "cuda" if torch.cuda.is_available() else "cpu"

video_path = "/content/fish.mp4"
output_video_path = "/content/segmented_video.mp4"

#Video info
cap = cv2.VideoCapture(video_path)
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps    = cap.get(cv2.CAP_PROP_FPS)
cap.release()
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))


results = model.predict(source=video_path, device=device, stream=True)

for r in results:
    out.write(r.plot()[:, :, ::-1])

out.release()
print("✅ Video saved to:", output_video_path)
```
